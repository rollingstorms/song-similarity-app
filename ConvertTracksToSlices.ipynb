```python
from pyarrow import feather
import os
```

```python
autoencoder_path = 'data/autoencoder_256dim_time_freq_128k_20'
tracks = feather.read_feather(autoencoder_path + '/tracks.feather')
```

```python
len(tracks)
```

```python
num_slices = 8
slices = len(tracks) // num_slices + 1
```

```python
tracks_length = len(tracks)
tracks_length - num_slices * slices
```

```python
slices
```

```python
tracks_slices = []
tracks_folder = autoencoder_path + '/tracks'
try:
    os.mkdir(tracks_folder)
except:
    pass
for num in range(num_slices):
    tracks_slice = tracks.iloc[num*slices : num*slices + slices].copy()
    feather.write_feather(tracks_slice, tracks_folder + f'/tracks_{num}.feather')
```

```python
tracks_folder_list
```

```python
import pandas as pd
tracks_folder_list = os.listdir(tracks_folder)
tracks = pd.DataFrame()
for track in tracks_folder_list:
    tracks = pd.concat([tracks, feather.read_feather(tracks_folder + "/" + track)])
tracks = tracks.sort_index()
```
