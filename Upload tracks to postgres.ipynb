```python
import json
from sqlalchemy import create_engine
from pyarrow import feather
import pandas as pd
import time
```

```python
#import secret url

f = open('.secrets/sql')
secrets = json.load(f)
url = secrets['url']
```

```python
#import tracks

tracks = feather.read_feather('test_128dim_newmels/tracks.feather')
tracks.drop(columns=['artist_genres'], inplace=True)
```

```python
#create sql engine

engine = create_engine(url, echo = False)
```

```python
#upload df to sql db

tracks.to_sql('tracks', con = engine, if_exists='append')
```

```python
#connect to python and test

import psycopg2
import os
con = psycopg2.connect(url)
cur = con.cursor()
```

```python
latent_cols = [f'latent_{i}' for i in range(128)]
```

```python
latent_cols_string = str(latent_cols).replace('[','').replace(']','')
```

```python
start = time.time()
results = pd.read_sql(f'SELECT {latent_cols_string} as latents FROM tracks', con)
print(time.time() - start)
```
